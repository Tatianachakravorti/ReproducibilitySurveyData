```python
import numpy as np
import matplotlib.pyplot as plt

color_map_blue = plt.cm.get_cmap('Blues')

labels = ['USA Eng', 'USA SS', 'India Eng', 'India SS']
counts = {
'quiet a lot' : [40.27777778, 75.26315789, 37.40458015, 36.84210526],
'a little bit' : [41.66666667, 19.47368421, 45.80152672, 42.10526316],
'not really' : [18.05555556, 5.263157895, 16.79389313, 21.05263158]
}
fig, ax = plt.subplots(figsize=(8, 8))
width = 0.9
bottom = np.zeros(4)
i = 1
for label, val in counts.items():
    p = ax.bar(x=labels, 
               height=np.round(val, decimals=2), 
               width=width, 
               label=label, 
               bottom=bottom, 
               color=color_map_blue(1 - 0.25*i),
               edgecolor='black')
    bottom += val
    i = i + 1
    ax.bar_label(p, label_type='center', fontsize=22)
ax.set_ylabel('Percentage', fontsize=20)
ax.legend(ncols=3, fontsize="15")
ax.spines[['right', 'top']].set_visible(False)
plt.ylim(0, 115)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```

```python
from matplotlib.pyplot import figure
import seaborn as sns
figure(figsize=(25, 5))
factors = ['Selective\nreporting', 'Pressure to\nPublish', 'Raw Data\nnot available', 'Code\nunavailable',
           'Insufficient\nPeer Review', 'Low Statistical\nPower']

N = len(factors) 
ind = np.arange(N) * 2.5
width = 0.5

india_sc = np.array([52.63, 44.73, 55.26, 28.94, 23.68, 23.68])
india_engg = np.array([44.27, 46.56, 59.54, 54.19, 32.82, 13.74])
usa_sc = np.array([67.37, 62.63, 51.58, 33.68, 14.74, 42.63])
usa_engg = np.array([58.33, 44.44, 65.28, 58.33, 27.78, 25])

bar1 = plt.bar(ind - width*1.5, india_engg, width, color = '#069002') 
bar2 = plt.bar(ind - width/2, india_sc, width, color='#69bc67')  
bar3 = plt.bar(ind + width/2, usa_engg, width, color = '#2067bf') 
bar4 = plt.bar(ind + width*1.5, usa_sc, width, color='#79a3d8')

plt.ylabel('Percentage', size=23) 
plt.yticks(size=23)
plt.xticks(ind,factors, size=23) 
sns.despine()
plt.legend((bar1, bar2, bar3, bar4), 
           ('India Eng', 'India SS', 'USA Eng', 'USA SS'),
           prop = {"size": 24},
           bbox_to_anchor=(1, 1), loc='upper left'
          )
plt.show() 
```
